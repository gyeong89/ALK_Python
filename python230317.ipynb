## 데이터 핸들링
- 행과 열을 다루는 방법

```python
lemonade.head()
```

- 새로운 칼럼 추가

```python
lemonade['Sold'] = 0
lemonade.head()
```

```python
lemonade['Sold'] = lemonade['Lemon'] + lemonade['Orange']
lemonade.head(3)
```

- Revenue 칼럼 : 매출 계산

```python
lemonade['Revenue'] = lemonade['Sold'] * lemonade['Price']
lemonade.head(3)
```

## 특정 칼럼 삭제


```python
# axis = 1, 컬럼을 삭제하겠다
lemonade_col_drop = lemonade.drop('Sold', axis = 1)
lemonade_col_drop.head(3)
```

- 행과 열을 동시 처리 --> loc만 처리

```python
#                         행 추출  조건식           열추출
result = lemonade.loc[lemonade['Revenue'] > 100, ['Date', 'Revenue']].reset_index(drop=True)
result
```

```python
# cols = ['Date','Revenue', 'Price']
result = lemonade.head(5).loc[lemonade['Sold'] > 50, ['Date', 'Lemon','Orange', 'Sold']].reset_index(drop=True)
result
```

## 집계함수(Aggregation)
- 그룹바이 및 집계 --> 피벗테이블

```python
import numpy as np
lemonade.groupby('Location')['Revenue'].agg([max, min, np.mean])
```

- 문제
- species 칼럼이 setosa인 것만 전체조회

```python
iris.head()
```

```python
result = iris[iris['species'] == 'setosa']
result.head(3)
```

- sepal_width 3보다 작은 데이터만 조회

```python
result2 = result.loc[result['sepal_width'] < 3].reset_index(drop=True)
result2
```

- AND 조건 &
- OR 조건 |

```python
result3 = iris[(iris['species'] == 'setosa') & (iris['sepal_width'] <3 )].reset_index(drop=True)
result3
```
